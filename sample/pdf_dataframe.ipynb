```python
import pandas as pd
from pathlib import Path

pdf_files = list(Path(".").glob("*.pdf"))

print("PDF files found:", len(pdf_files))
pdf_files
```

```python
from pdf_extractor import extract_all_attributes, extract_current_directors, extract_company_name, extract_year_ended

rows = []

for pdf in pdf_files:
    print("Processing:", pdf.name)

    data = extract_all_attributes(pdf)

    directors = extract_current_directors(data)
    company_name = extract_company_name(data)
    year_ended = extract_year_ended(data)

    row = {
        "file": pdf.name,
        "company_name": company_name,
        "year_ended": year_ended,
        "page_count": data["metadata"]["page_count"],
        "num_images": len(data["images"]),
        "num_links": len(data["links"]),
        "num_annotations": len(data["annotations"]),
        "num_fonts": len(data["fonts"]),
        "current_directors": directors
    }

    rows.append(row)

df = pd.DataFrame(rows)
df
```
